```python
sns.set(style='darkgrid')
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
sns.histplot(data=df, x=df['SepalLengthCm'], kde=True, ax=ax[0,0], hue='Species')
sns.histplot(data=df, x=df['SepalWidthCm'], kde=True, ax=ax[0,1], hue='Species')
sns.histplot(data=df, x=df['PetalLengthCm'], kde=True, ax=ax[1,0], hue='Species')
sns.histplot(data=df, x=df['PetalWidthCm'], kde=True, ax=ax[1,1], hue='Species')
```
