```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs = pd.read_csv('cluster_blobs.csv')
blobs.head(2)
```

```python
sns.scatterplot(data=blobs, x='X1', y='X2')
plt.show()
```

```python
moons = pd.read_csv('cluster_moons.csv')
moons.head(2)
```

```python
sns.scatterplot(data=moons, x='X1', y='X2')
```

```python
circles = pd.read_csv('cluster_circles.csv')
circles.head(2)
```

```python
sns.scatterplot(data=circles, x="X1", y="X2")
```

```python
def display_categories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1', y='X2', hue=labels, palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
display_categories(model, blobs)
```

```python
model = KMeans(n_clusters=2)
display_categories(model, circles)
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN()
display_categories(model, blobs)
```

```python
model = DBSCAN(eps=0.15)
display_categories(model, moons)
```

```python
model = DBSCAN(eps=0.10)
display_categories(model, circles)
```

```python
def display_categories2(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1', y='X2', hue=labels, palette='Set1')
```

```python
blob1 = DBSCAN()
display_categories2(blob1, two_blobs)
```

```python
blob1_outliers = DBSCAN()
display_categories2(blob1_outliers, two_blobs_outliers)
```

```python
blob1_outliers.labels_
```

### Determine the optimal EPS value for DBScan

```python
outlier_percent = []
num_of_outliers = []

for eps in np.linspace(0.01, 10, 100):
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(two_blobs_outliers)
    
    #total nummber of outliers found
    num_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    perc_outliers = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
    outlier_percent.append(perc_outliers)
```

```python
sns.lineplot(x=np.linspace(0.01, 10, 100), y=num_of_outliers)
plt.xlabel('EPS value')
plt.xlim(0,0.4)
```
