Select the best features in the Wine dataset.

1. Look for candidate-features that have a pearson correlation coefficient less than -0.5 and more than 0.5 with respect to "Cultivar" (our target).

2. For each candidate-feature, extract peers that have a coefficient more than 0.6 with itself. Then, among that candidate-feature and its highly-correlated peers, select the one that has the highest coefficient value with resepect to "Cultivar". Discard the rest of the unselected features and do not consider them in future iterations.

3. Iterate until all candidate-features (that have not been discarded in earlier iterations) have been processed.

4. The selected ones in each iteration are the "best features".

Perform Step 1 below.

```python
series = corr_matrix['Cultivar'].drop('Cultivar')
list_of_correlated_columns = []
for i in range(len(series)):
    if series[i] < -0.5 or series[i] > 0.5:
        list_of_correlated_columns.append(series.index[i])
# or you can use boolean indexing
# list_of_correlated_columns = series[(series < -0.5) | (series > 0.5)]
```

```python
skip = []
selected = []

for x in list_of_correlated_columns:
    if x not in skip and x not in selected:
        features_wrt_x = corr_matrix[x].drop(['Cultivar']) # all known features related to x
        print(type(features_wrt_x))
        peers = features_wrt_x[(list_of_correlated_columns)]
        peers = peers[(peers > 0.6)] # remaining list of features that are related
        print(type(peers))
        if len(peers) == 1:
            selected.append(x)
        else:
            # select the most correlated features to target because there are at least 
            # one feature (not itself) that could be correlated to target
            series_wrt_target = corr_matrix['Cultivar'][peers.index]
            max_corr_feature = max(series_wrt_target)
            for i in range(len(corr_matrix['Cultivar'])):
                if max_corr_feature == corr_matrix['Cultivar'][i]:
                    feature_to_add = corr_matrix['Cultivar'].index[i]
                    selected.append(feature_to_add)
                    series_wrt_target.drop(feature_to_add, inplace=True)
                    # place unselected peers into skip
                    skip.extend(series_wrt_target)
                    break
            
            
            
            
            
```
