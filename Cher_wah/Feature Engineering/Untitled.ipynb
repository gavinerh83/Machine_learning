```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head(2)
df['Species'].value_counts()
```

```python
df.drop(['Id'], axis=1, inplace=True)
```

```python
sns.set(style='darkgrid')
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

sns.kdeplot(data=df, x='SepalLengthCm', hue='Species', ax=ax[0,0])
sns.kdeplot(data=df, x='SepalWidthCm', hue='Species', ax=ax[0,1])
sns.kdeplot(data=df, x='PetalLengthCm', hue='Species', ax=ax[1,0])
sns.kdeplot(data=df, x='PetalWidthCm', hue='Species', ax=ax[1,1])
plt.show()
```

```python
x = df.iloc[:,0:-1].values
```

```python
y = df.loc[:,'Species'].values
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
sns.set(style='darkgrid')
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

label_dict = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
k=0
for i in range(2):
    for j in range(2):
        sns.kdeplot(data=x_scaled[:50,k], ax=ax[i,j])
        sns.kdeplot(data=x_scaled, x=x_scaled[50:100, k], ax=ax[i,j])
        sns.kdeplot(data=x_scaled, x=x_scaled[100:150, k], ax=ax[i,j])
        ax[i,j].set_xlabel(label_dict[k])
        k += 1
        
plt.show()
```

```python
pca = PCA(n_components=2)

pc = pca.fit_transform(x_scaled)
```
