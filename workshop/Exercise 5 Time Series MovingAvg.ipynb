# Time Series - Moving Average Exercise 2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AMZN.csv", index_col=0, parse_dates=True, infer_datetime_format=True)
df.head()
```

```python
df.index.dtype
```

```python
rolling_50 = df['ClosingPrice'].rolling(window=50).mean().shift(1)
rolling_200 = df['ClosingPrice'].rolling(window=200).mean().shift(1)
```

```python
plt.plot(df.index, df.ClosingPrice, label='Actual Closing Price')
plt.plot(df.index, rolling_50, label='50 Day moving Average')
plt.plot(df.index, rolling_200, label='200 Day moving average')
plt.legend(loc='upper left')
plt.show()
```

### Mean square error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(df.ClosingPrice[50:], rolling_50[50:])
```

```python
mean_squared_error(df.ClosingPrice[200:], rolling_200[200:])
```
